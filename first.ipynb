## Aggregation of data

```python
washersdata[["BrandName"]].describe()
```

```python
washersdata[["Volume"]].describe() ## for integer feature
```

```python
washersdata[["BrandName"]].value_counts()
```

```python
washersdata[["BrandName"]].value_counts(normalize = True)
```

```python
washersdata[["Volume"]].mean()
```

```python
washersdata.groupby("BrandName")[["Volume"]].mean()
```

```python
washersdata.groupby("BrandName")[["Volume"]].mean().sort_values(by = "Volume")
```

```python
washersdata.groupby("BrandName")[["Volume"]].agg(["mean", "median", "min", "max"])
```
